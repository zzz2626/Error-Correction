```python
#Bar chart without the black line
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("TechGadgetSales.csv")

plt.figure(figsize=(8, 6))
sns.set_theme(style='whitegrid', palette='muted')

barplot = sns.barplot(data=data, x='Income', y='MaritalStatus', hue='Gender', palette='ch:30',errorbar=None)

for container in barplot.containers:
    barplot.bar_label(container, fmt='%.0f', label_type='edge', padding=3)  

plt.xlabel('Income')
plt.ylabel('Marital Status')
plt.title('Horizontal Barplot of Income by Gender and Marital Status')
plt.legend(title='Gender')
plt.show()
```

```python
#Bar chart with the black line
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("TechGadgetSales.csv")

plt.figure(figsize=(8, 6))
sns.set_theme(style='whitegrid', palette='muted')

barplot = sns.barplot(data=data, x='Income', y='MaritalStatus', hue='Gender', palette='ch:30') #By default, it is aggregated by the mean

for container in barplot.containers:
    barplot.bar_label(container, fmt='%.0f', label_type='edge', padding=3)  

plt.xlabel('Income')
plt.ylabel('Marital Status')
plt.title('Horizontal Barplot of Income by Gender and Marital Status')
plt.legend(title='Gender')
plt.show()
```
